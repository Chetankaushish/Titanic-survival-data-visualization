```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Cabin']=df['Cabin'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Embarked'])
```

```python
df.isnull().sum()
```

```python
#Heatmap of Age vs Gender
pivot=df.pivot_table(index='Sex',columns='Age',aggfunc='size')
sns.heatmap(pivot,cmap='YlGnBu')
plt.title('Heatmap of Age vs Gender')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.show()
#Heatmap shows relationship b/w Age and Gender passenger Travelled on the Titanic 
```

```python
#Pairplot for showing Survived Data
sns.pairplot(df, hue="Survived", diag_kind="kde")

plt.show()
#In this pairplot shows total passenger Survived according to their PassengerId,Pclass,Age,SibSp,Parch,Fare
```

```python
sns.boxplot(x="Pclass", y="Age",hue='Sex', data=df, palette="Set2")

plt.show()
```

```python
plt.scatter(df["Age"], df["Survived"], marker='o', color='blue')
plt.xlabel("Age")
plt.ylabel("Total alive")
plt.title("Passenger Survival Scatter Plot")
plt.show()
```

```python
plt.hist(df["Age"], bins=10, color='teal', edgecolor='black')
plt.title("Age Distribution Histogram")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
#




#From the histogram, we observe that younger passengers, especially children, had higher survival rates, while middle-aged and elderly passengers were less likely to survive. This indicates that priority was given to children during evacuation, aligning with the historical context of the Titanic disaster.
```
